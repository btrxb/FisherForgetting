```python
alpha_list = [1e-7, 2.5e-7, 5e-7, 7.5e-7, 1e-06]
forget_alpha = []
retain_alpha = []
test_alpha = []

for alpha in alpha_list:
    forget_alpha.append(pd.read_csv(f'acc0_forget_{alpha}.csv', index_col=0))
    retain_alpha.append(pd.read_csv(f'acc0_{alpha}.csv', index_col=0))
    test_alpha.append(pd.read_csv(f'acc0_test_{alpha}.csv', index_col=0))

forget_alpha_df = pd.concat(forget_alpha)
retain_alpha_df = pd.concat(retain_alpha)
test_alpha_df = pd.concat(test_alpha)

forget_alpha_df = forget_alpha_df.reset_index(drop=True)
retain_alpha_df = retain_alpha_df.reset_index(drop=True)
test_alpha_df = test_alpha_df.reset_index(drop=True)
```

```python
forget_alpha_df.index = alpha_list
retain_alpha_df.index = alpha_list
test_alpha_df.index = alpha_list
```

```python
column_names = ['Eredeti', 'Finomhangolt', 'Fisher', 'Újratanított']
forget_alpha_df.columns = column_names
retain_alpha_df.columns = column_names
test_alpha_df.columns = column_names
```

```python
forget_alpha_df = forget_alpha_df*100
retain_alpha_df = retain_alpha_df*100
test_alpha_df = test_alpha_df*100
```
